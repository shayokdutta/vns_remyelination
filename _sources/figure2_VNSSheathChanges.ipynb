# Figure 2 VNS Alone Does Not Affect Sheath Replacement/Patterns

## Panels A & B: Categorizing Myelin Regeneration
**Quality over Quantity.** While Figure 1 established that VNS increases the *amount* of regeneration, Figure 2 asks whether it restores the original *pattern* of myelination.

![Figure 2 Panel AB](figure2/panelAB.svg)
<div class="alert alert-block alert-info">
    <b>Methodological Context: Sheath Classification</b><br>
    To determine if VNS guides repair to the correct locations, we utilized high-resolution longitudinal tracing to categorize every new myelin sheath (<i>Fig. 2b</i>):
    <ul>
        <li><b>"New" (De Novo):</b> Sheaths generated in locations that were previously <i>unmyelinated</i>. This represents a change in the circuit topology.</li>
        <li><b>"Replaced" (Restored):</b> Sheaths generated in locations that were <i>previously myelinated</i> and subsequently lost during cuprizone treatment. This represents true repair of the original circuit map.</li>
        <li><b>The Biological Question:</b> Does VNS specifically enhance the restoration of the original map ("Replaced"), or does it simply drive broad, non-specific myelination ("New")?</li>
    </ul>
</div>

## Panel C -- E: OL Morphology & Targeting
![Figure 2 Panel CDE](figure2/panelCDE.svg)

```python
import pandas as pd
from IPython.display import display, HTML

# Formatting function
def pretty_print_stats(df, title):
    print(f"\n{title}")
    display(HTML(df.to_html(index=False, classes='table table-striped table-hover', justify='left')))

# --- PANELS C, D, E DATA (Cellular Morphology & Targeting) ---
data_morph = {
    "Metric": [
        "Sheath Number per Oligodendrocyte (Fig. 2C)", 
        "Average Sheath Length (Fig. 2D)", 
        "% Sheath Replacement (Targeting Accuracy) (Fig. 2E)"
    ],
    "VNS Mean ± SEM": [
        "50.17 ± 2.67", 
        "56.87 ± 3.28 µm", 
        "20.71 ± 3.11 %"
    ],
    "Control Mean ± SEM": [
        "44.47 ± 2.88", 
        "63.02 ± 3.11 µm", 
        "30.84 ± 5.20 %"
    ],
    "Statistical Test": [
        "Two-tailed t-test", 
        "Two-tailed t-test", 
        "Two-tailed t-test"
    ],
    "P-Value": [
        "0.1874 (n.s.)", 
        "0.2081 (n.s.)", 
        "0.1481 (n.s.)"
    ],
    "Variance Equality (Brown-Forsythe)": [
        "p=0.3987 (Equal)", 
        "p=0.9173 (Equal)", 
        "p=0.1717 (Equal)"
    ]
}
df_morph = pd.DataFrame(data_morph)

# Display
pretty_print_stats(df_morph, "--- Panels C-E Statistics: Cellular Morphology & Targeting ---")
```

<div class="alert alert-block alert-success">
    <b>Statistical Insight: "More Cells, Same Rules"</b><br>
    These three null results provide a crucial mechanistic constraint on how VNS operates:
    <ul>
        <li><b>Preserved Morphology (Panels C & D):</b> VNS did not alter the number of sheaths formed per cell (<i>p = 0.1874</i>) or their average length (<i>p = 0.2081</i>). This implies that VNS does not "supercharge" individual oligodendrocytes to grow larger or more complex; it simply recruits <b>more of them</b>.</li>
        <li><b>No Targeting Bias (Panel E):</b> The "Targeting Accuracy" (% of sheaths that replaced lost myelin) was statistically indistinguishable between groups (<i>p = 0.1481</i>). VNS does not appear to possess a specific guidance cue that directs new myelin to previously injured axons over naive ones.</li>
        <li><b>Conclusion:</b> The therapeutic effect is driven by <b>population expansion</b> (Quantitative), not by altering single-cell biology or guidance mechanisms (Qualitative).</li>
    </ul>
</div>

## Panels F -- H: Mapping the Global Myelin Landscape
**Volumetric Tracing.** To capture the complete impact of VNS on circuit topology, we reconstructed the entire myelin sheath population within a defined $150 \times 150 \times 60 \mu m^3$ cortical volume at both Baseline (-21 Days) and Recovery (+31 Days).

![Figure 2 Panel FGH](figure2/panelFGH.svg)

<div class="alert alert-block alert-info">
    <b>Methodological Context: The "Myelin Fingerprint"</b><br>
    By overlaying the myelin maps from before and after the injury (<i>Fig. 2f, h</i>), we classified every sheath into three distinct topological categories:
    <ul>
        <li><b>Lost (Magenta):</b> Sheaths present at baseline but absent after recovery. These represent permanent circuit deficits.</li>
        <li><b>Survived / Replaced (Grey):</b> Sheaths present at both timepoints. This category includes original sheaths that survived the injury and lost sheaths that were precisely restored in the same location.</li>
        <li><b>De Novo (Cyan):</b> Sheaths absent at baseline but present after recovery. These represent <i>new</i> myelination patterns that alter the original circuit topology.</li>
    </ul>
</div>

## Panels I -- L: VNS Drives De Novo Myelination, Not Pattern Restoration

![Figure 2 Panel IJKL](figure2/panelIJKL.svg)

```python
# --- PANELS I, J, K DATA (Pattern Restoration) ---
data_topo = {
    "Metric": [
        "Replaced Sheath Count (Absolute) (Fig. 2I)", 
        "Survived Sheath Count (Absolute) (Fig. 2I)", 
        "% Sheath Replacement (Relative Fidelity) (Fig. 2J)", 
        "% Pattern Similarity (Global Match) (Fig. 2K)"
    ],
    "VNS Mean ± SEM": [
        "41.13 ± 3.32", 
        "43.98 ± 7.14", 
        "17.74 ± 6.16 %", 
        "35.83 ± 6.39 %"
    ],
    "Control Mean ± SEM": [
        "25.54 ± 0.63", 
        "21.22 ± 7.32", 
        "46.46 ± 5.94 %", 
        "59.01 ± 6.16 %"
    ],
    "Statistical Test": [
        "REML Mixed Model (F-Test)", 
        "REML Mixed Model (F-Test)", 
        "REML Mixed Model (F-Test)", 
        "REML Mixed Model (F-Test)"
    ],
    "P-Value": [
        "0.0705 (Trend)", 
        "0.4798 (n.s.)", 
        "0.0612 (Trend)", 
        "0.1291 (n.s.)"
    ],
    "Interpretation": [
        "VNS trends higher (Absolute count)", 
        "No difference in survival", 
        "VNS trends LOWER (Diluted by De Novo)", 
        "VNS trends LOWER (Altered Map)"
    ]
}
df_topo = pd.DataFrame(data_topo)

# Display
pretty_print_stats(df_topo, "--- Panels I-L Statistics: Pattern Restoration & Fidelity ---")
```

<div class="alert alert-block alert-warning">
    <b>Statistical Insight: The "Dilution" of the Original Map</b><br>
    These panels reveal a critical limitation of VNS-driven repair—it is additive, not restorative:
    <ul>
        <li><b>Absolute vs. Relative (The Paradox):</b> While VNS trended toward increasing the <i>absolute number</i> of replaced sheaths (41 vs 25, <i>p=0.07</i>), it actually trended toward a <i>lower percentage</i> of replacement fidelity (17% vs 46%, <i>p=0.06</i>).</li>
        <li><b>Mechanism:</b> This discrepancy is driven by the massive influx of <b>De Novo</b> myelin (previously unmyelinated locations). VNS floods the circuit with new myelin, which "dilutes" the proportion of sheaths that land in the correct original locations.</li>
        <li><b>Conclusion:</b> VNS does not strictly restore the native myelin pattern; instead, it creates a new, hybrid circuit topology dominated by de novo sheaths.</li>
    </ul>
</div>
