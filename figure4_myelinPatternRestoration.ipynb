# Figure 4 Paired VNS Restores Native Myelin Pattern

## Panels A & B: New OL Sheath Counts & Lengths
**Quantity vs. Quality.** Before assessing *where* the myelin goes, we established whether Paired VNS alters the intrinsic properties of the new oligodendrocytes.

![Figure 4 Panel AB](figure4/panelAB.svg)

<div class="alert alert-block alert-info">
    <b>Experimental Insight: Targeting, Not Capacity</b><br>
    Similar to our findings with open-loop VNS (Fig. 2), pairing VNS with motor learning did not fundamentally alter the morphology of individual cells:
    <ul>
        <li><b>Sheath Capacity (Fig. 4a):</b> New oligodendrocytes generated the same number of sheaths regardless of the intervention (<i>p = 0.1758</i>).</li>
        <li><b>Sheath Length (Fig. 4b):</b> The average length of these sheaths was statistically equivalent between groups (<i>p = 0.0793</i>).</li>
    </ul>
    <b>Conclusion:</b> This confirms that the improved functional outcomes are not due to individual cells becoming "super-myelinators." Instead, it points to a specific change in <b>where</b> these cells are placing their myelin.
</div>

## Panel C: Sheath Replacement
![Figure 4 Panel C](figure4/panelC.svg)

```python
import pandas as pd
from IPython.display import display, HTML

# Formatting function
def pretty_print_stats(df, title):
    print(f"\n{title}")
    display(HTML(df.to_html(index=False, classes='table table-striped table-hover', justify='left')))

# --- PANEL C DATA (Targeting Accuracy) ---
data_c = {
    "Comparison": ["Paired VNS vs. Motor Learning"],
    "Metric": ["% Sheath Replacement (Targeting Accuracy)"],
    "Paired VNS Mean ± SEM": ["47.46 ± 4.41 %"],
    "Motor Learning Mean ± SEM": ["29.02 ± 4.80 %"],
    "Test": ["Two-tailed t-test"],
    "P-Value": ["0.0213 (*)"],
    "Variance Equality (Brown-Forsythe)": ["p=0.2381 (Equal)"]
}
df_c = pd.DataFrame(data_c)

# Display
pretty_print_stats(df_c, "--- Panel C Statistics: The \"Targeting\" Shift ---")
```

<div class="alert alert-block alert-success">
    <b>Statistical Insight: Activity-Dependent Targeting</b><br>
    This result marks the critical divergence between open-loop (Fig. 2) and closed-loop stimulation:
    <ul>
        <li><b>The Failure of Open-Loop:</b> Recall that open-loop VNS <a href="figure2_VNSSheathChanges.html">did not improve the targeting of new sheaths (Fig. 2e, p=0.1481)</a>.</li>
        <li><b>The Success of Pairing:</b> By synchronizing VNS with successful motor reaches, we observed a significant increase in <b>Sheath Replacement</b> (<i>p=0.0213</i>). Nearly <b>50%</b> of all new sheaths in the Paired VNS group were directed to repair previously myelinated axons, compared to only ~29% in the Motor Learning controls.</li>
        <li><b>Mechanism:</b> This suggests that timing VNS to circuit activation provides the necessary "spatial tag" to guide new myelin to the specific axons that require metabolic support for the task.</li>
    </ul>
</div>

## Panels E & F: Specificity of Repair

![Figure 4 Panel EF](figure4/panelEF.svg)

```python
import pandas as pd
from IPython.display import display, HTML

# Formatting function
def pretty_print_stats(df, title):
    print(f"\n{title}")
    # Highlight significant p-values
    def highlight_sig(val):
        if isinstance(val, str) and ('*' in val or '<' in val) and not ('n.s.' in val):
            return 'background-color: #d4edda; font-weight: bold;'
        return ''

    styled_df = df.style.applymap(highlight_sig, subset=['P-Value']).set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'left')]}]
    ).hide(axis='index')
    
    display(HTML(styled_df.to_html(classes='table table-striped table-hover', index=False)))

# --- PANELS E & F DATA (Survival vs Replacement) ---
data_ef = {
    "Metric": [
        "Survived Sheath Count (Fig. 4E)",
        "Replaced Sheath Count (Fig. 4E)",
        "% Myelin Replacement (Relative Yield) (Fig. 4F)"
    ],
    "Paired VNS Mean ± SEM": [
        "33.19 ± 5.31", 
        "48.79 ± 2.93", 
        "35.19 ± 1.81 %"
    ],
    "Motor Learning Mean ± SEM": [
        "27.57 ± 3.35", 
        "29.81 ± 1.92", 
        "23.79 ± 1.07 %"
    ],
    "Statistical Test": [
        "REML Mixed Model (F-Test)", 
        "REML Mixed Model (F-Test)", 
        "REML Mixed Model (F-Test)"
    ],
    "P-Value": [
        "0.5938 (n.s.)", 
        "0.0022 (**)", 
        "0.0098 (**)"
    ],
    "Interpretation": [
        "VALIDATION: Identical baseline injury", 
        "EFFECT: VNS restores lost sheaths", 
        "EFFECT: VNS improves repair efficiency"
    ]
}
df_ef = pd.DataFrame(data_ef)

# Display
pretty_print_stats(df_ef, "--- Panels E & F Statistics: Specificity of Repair ---")
```

<div class="alert alert-block alert-success">
    <b>Statistical Insight: Identical Baseline, Divergent Recovery</b><br>
    The decomposition of sheath fate validates our experimental timeline and highlights the specificity of the repair:
    <ul>
        <li><b>Validating the Baseline (Survival):</b> Since the intervention began <i>after</i> the demyelination period, we expected no difference in sheath survival. The statistical equivalence (<i>p = 0.5938</i>) confirms that both groups entered the recovery phase with <b>identical structural deficits</b>, ruling out sampling bias or unequal injury.</li>
        <li><b>Targeted Restoration (Replacement):</b> With the baseline confirmed, the divergence in replacement (<i>p = 0.0022</i>) indicates that Paired VNS actively recruits new oligodendrocytes to fill these pre-existing gaps.</li>
        <li><b>The Efficiency Shift (Panel F):</b> Crucially, Paired VNS significantly increased the <i>efficiency</i> of this repair (<i>p = 0.0098</i>), reversing the trend seen in open-loop stimulation (<a href="figure2_VNSSheathChanges.html">Fig. 2j</a>) where replacement efficiency actually trended downward.</li>
    </ul>
</div>

## Panels G & H: Paired VNS Restores Native Myelin Pattern!
![Figure 4 Panel GH](figure4/panelGH.svg)

```python
import pandas as pd
from IPython.display import display, HTML

# Formatting function
def pretty_print_stats(df, title):
    print(f"\n{title}")
    # Highlight significant p-values
    def highlight_sig(val):
        if isinstance(val, str) and ('*' in val or '<' in val):
            return 'background-color: #d4edda; font-weight: bold;'
        return ''

    styled_df = df.style.applymap(highlight_sig, subset=['P-Value']).set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'left')]}]
    ).hide(axis='index')
    
    display(HTML(styled_df.to_html(classes='table table-striped table-hover', index=False)))

# --- PANELS G & H DATA (Replacement Efficiency vs. Pattern Restoration) ---
data_gh = {
    "Comparison": ["Paired VNS vs. Motor Learning", "Paired VNS vs. Motor Learning"],
    "Metric": [
        "Panel G: % Myelin Sheath Replacement (Efficiency)", 
        "Panel H: % Pattern Restoration (Final Topology)"
    ],
    "Paired VNS Mean ± SEM": [
        "32.91 ± 3.07 %", 
        "47.64 ± 3.54 %"
    ],
    "Motor Learning Mean ± SEM": [
        "22.97 ± 2.20 %", 
        "37.55 ± 1.84 %"
    ],
    "Test": ["Student's t-test", "Student's t-test"],
    "P-Value": ["0.0179 (*)", "0.0142 (*)"],
    "Conclusion": [
        "Significantly higher efficiency in replacing lost sheaths", 
        "Significantly higher fidelity to the original circuit map"
    ]
}
df_gh = pd.DataFrame(data_gh)

# Display
pretty_print_stats(df_gh, "--- Panels G & H Statistics: The Mechanism and The Outcome ---")
```

<div class="alert alert-block alert-success">
    <b>The Grand Conclusion: True Circuit Restoration</b><br>
    These two panels provide the definitive evidence that Paired VNS drives targeted repair rather than random regeneration:
    <ul>
        <li><b>The Process (Fig. 4g):</b> Paired VNS significantly increased the <b>efficiency of replacement</b> (<i>32.9% vs 23.0%, p=0.0179</i>). This confirms that when the brain builds new myelin under closed-loop stimulation, it preferentially targets the axons that lost myelin.</li>
        <li><b>The Outcome (Fig. 4h):</b> Crucially, this efficient replacement culminated in a restored circuit topology. The <b>Pattern Restoration</b> metric—which measures how closely the final myelin map resembles the pre-injury baseline—was significantly higher in the Paired VNS group (<i>47.6% vs 37.6%, p=0.0142</i>).</li>
        <li><b>Synthesis:</b> While Motor Learning alone leaves the circuit permanently altered, adding VNS restores the native myelin pattern, effectively "healing" the map.</li>
    </ul>
</div>
